```python
#Author: Victor Antonio de Oliveira
#nUSP: 9791326

#Assignment 2 : Image Enhancement and Filtering
import numpy as np
import imageio
import bf
import umlf
import vf
from matplotlib import pyplot as plt

input_img = imageio.imread("../images/camera.png")
input_img_array = np.array(input_img).astype(np.int32)
```

We'll be applying all three methods to the image *camera.png* with static values to its parameters. The values can be seen as predetermined numbers below.
After applying all methods into this image, we'll be applying into others as well.

```python

filter_size = 3
sigma_s = 150.0
sigma_r = 100.0

filter_matrix = bf.create_filter_matrix(filter_size)
spatial = bf.gaussian_kernel(filter_matrix, sigma_s)
output_img_array = bf.bilateral_filter(input_img_array, spatial, sigma_r)


output_img = output_img_array.astype(np.uint8)
plt.imshow(output_img, cmap='gray')
plt.show() 
```

```python
param_c = 0.5
k_number = 1

output_img_array = umlf.unsharp_mask(input_img_array, param_c, k_number)
output_img = output_img_array.astype(np.uint8) 
    
plt.imshow(output_img, cmap='gray')
plt.show() 
```

```python
sigma_row = 50.0
sigma_col = 50.0

output_img_array = vf.vignette_filter(input_img_array, sigma_row, sigma_col)
output_img = output_img_array.astype(np.uint8) 

plt.imshow(output_img, cmap='gray')
plt.show() 
```
